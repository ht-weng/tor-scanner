```python
import pandas as pd
```

```python
def count_open(series):
    return series.equals("open").sum(1)
```

```python
url_ports = pd.read_csv("../data/onions_ports.csv")
url_ports
```

```python
url_ports["open_ports"] = url_ports.isin({"open"}).sum(1)
url_ports.sort_values(by=["open_ports"], ascending=False, inplace=True)
url_ports
```

```python
url_ports_grouped_by_count = url_ports[["url", "open_ports"]].groupby(["open_ports"]).count()
url_ports_grouped_by_count.rename(columns={"url": "count"}, inplace=True)
url_ports_grouped_by_count.reset_index(inplace=True)
url_ports_grouped_by_count.sort_values(by=["open_ports"], ascending=False, inplace=True)
url_ports_grouped_by_count
```

```python
port_count = []
port_list = ["20", "21", "22", "23", "25", "26", "80", "110", "143", "443", "587", "993", "995", "2525", "3306"]
for port in port_list:
    count = len(url_ports[url_ports[port] == "open"])
    port_count.append(count)

port_count_df = pd.DataFrame([port_list, port_count])
port_count_df = port_count_df.transpose()
port_count_df.rename(columns={0: "port", 1: "count_open"}, inplace=True)
port_count_df.sort_values(by=["count_open"], ascending=False, inplace=True)
port_count_df
```
